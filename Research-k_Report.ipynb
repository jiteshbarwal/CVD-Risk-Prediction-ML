# FEATURE SELECTION -> Singular value decomposition on k nearest   neighbour

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import TruncatedSVD
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Assuming X is your feature matrix and y is your target variable
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Standardize the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Apply Singular Value Decomposition
n_components = 10  # Set the desired number of components
svd = TruncatedSVD(n_components=n_components)
X_train_svd = svd.fit_transform(X_train_scaled)
X_test_svd = svd.transform(X_test_scaled)

# Initialize and train the KNN classifier on the reduced feature space
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train_svd, y_train)

# Make predictions on the test set
y_pred = knn.predict(X_test_svd)

# Evaluate the performance
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy}')
```

# FEATURE SELECTION -> Singular value decomposition on naive boyes

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import TruncatedSVD
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

# Assuming X is your feature matrix and y is your target variable
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Standardize the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Apply Singular Value Decomposition
n_components = 10  # Set the desired number of components
svd = TruncatedSVD(n_components=n_components)
X_train_svd = svd.fit_transform(X_train_scaled)
X_test_svd = svd.transform(X_test_scaled)

# Initialize and train a Naive Bayes classifier on the reduced feature space
nb = GaussianNB()
nb.fit(X_train_svd, y_train)

# Get the feature probabilities from Naive Bayes
feature_probabilities = nb.theta_

# Map the probabilities to the original feature names
feature_names = [f'Feature_{i}' for i in range(X_train.shape[1])]
selected_feature_names_nb = [feature_names[i] for i in range(len(feature_probabilities[0]))]

# Print the names of the selected attributes after Naive Bayes
print(f'Selected Feature Names after Naive Bayes: {selected_feature_names_nb}')

# Make predictions on the test set
y_pred = nb.predict(X_test_svd)

# Evaluate the performance
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy}')
```

# FEATURE SELECTION -> Singular value decomposition on logistic regression

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import TruncatedSVD
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Assuming X is your feature matrix and y is your target variable
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Standardize the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Apply Singular Value Decomposition
n_components = 10  # Set the desired number of components
svd = TruncatedSVD(n_components=n_components)
X_train_svd = svd.fit_transform(X_train_scaled)
X_test_svd = svd.transform(X_test_scaled)

# Get the original feature names
feature_names = [f'Feature_{i}' for i in range(X_train.shape[1])]

# Print the names of the selected attributes after TruncatedSVD
print(f'Selected Feature Names after TruncatedSVD: {feature_names[:n_components]}')

# Initialize and train a Logistic Regression classifier on the reduced feature space
logreg = LogisticRegression()
logreg.fit(X_train_svd, y_train)

# Get the coefficients of logistic regression
logreg_coefficients = logreg.coef_

# Map the coefficients to the original feature names
selected_feature_names_logreg = [feature_names[i] for i in range(len(logreg_coefficients[0]))]

# Print the names of the selected attributes after Logistic Regression
print(f'Selected Feature Names after Logistic Regression: {selected_feature_names_logreg}')

# Make predictions on the test set
y_pred = logreg.predict(X_test_svd)

# Evaluate the performance
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy}')
```

# FEATURE SELECTION -> linear discriminant analysis on k nearest neighbour

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Assuming X is your feature matrix and y is your target variable
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Standardize the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Apply Linear Discriminant Analysis
lda = LinearDiscriminantAnalysis()
X_train_lda = lda.fit_transform(X_train_scaled, y_train)
X_test_lda = lda.transform(X_test_scaled)

# Get the coefficients of LDA transformation
lda_coefficients = lda.coef_

# Map the coefficients to the original feature names
feature_names = [f'Feature_{i}' for i in range(X_train.shape[1])]
selected_feature_names = [feature_names[i] for i in range(len(lda_coefficients[0]))]

# Print the names of the selected attributes
print(f'Selected Feature Names: {selected_feature_names}')

# Initialize and train the KNN classifier on the reduced feature space
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train_lda, y_train)

# Make predictions on the test set
y_pred = knn.predict(X_test_lda)

# Evaluate the performance
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy}')
```
